### Task 1.1 – Central Tendency

For **monthly revenue**:

* Calculate:
  * Mean
  * Median
  * Mode

```python
mean= df['revenue'].mean()
mean
```

```python
median = df['revenue'].median()
median
```

```python
df['revenue'].mode()
```

**which measure best represents revenue** and why (consider outliers).  
Median best represents revenue because it is not affected by outliers

### Task 1.2 – Dispersion

Calculate:

* Range
* Variance
* Standard deviation

```python
# Range = max(x) - min(x)
range = df['revenue'].max() - df['revenue'].min()
range
```

```python
# Variance [high variance means the data is very spread out; a low variance means the numbers are close to the mean.]
variance = df['revenue'].var()
variance
```

```python
# Standard Deviation [square root of the variance.]
print(df['revenue'].std())
```

* What does high standard deviation imply about sales stability?  
High standard deviation implies that sales are unstable

## PART 2: Data Visualization

Create:

* Line chart - Revenue over time
* Bar chart - Revenue by store type
* Box plot - Revenue distribution by region
* Scatter plot - Marketing spend vs revenue

```python
# Line Chart
plt.plot(df['date'], df['revenue'])
```

```python
# Bar Chart (Sorted)
df.groupby('store_type')['revenue'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Total Revenue by Store Type')
```

```python
# Box Plot
df.boxplot(column='revenue', by='region')
plt.title('Revenue Distribution by Region')
```

```python
# Scatter Plot
plt.scatter(df['units_sold'], df['revenue'])
plt.title('units_sold vs Revenue')
```

 **what each visualization reveals**.  
line chart shows trends how revenue is over time     
bar chart compares the online vs the physical store    
box plot shows the median and outliers for each region    
scatter plot looks for a correlation

### Task 6.1 – Errors


* Type I Error   
A false positive   
Saying there's an effect/difference when there actually isn't
* Type II Error
A false negative  
Saying there's no effect when there actually is one

Example:
* What would a Type I error mean for the company?  
A retail chain tests a new merchandise arrangement in 20 stores for 1 month.

Data shows: 15% revenue increase (statistically significant at α=0.05)

Reality: Random variation or seasonal effect, not the new arrangement

Type I Error: Company rolls it out to all 500 stores

Result: No actual benefit, but $2M spent on implementation
